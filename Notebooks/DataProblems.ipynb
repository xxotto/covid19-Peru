# Análisis Exploratorio de Data Product 6

Se tiene registro [(Data Product 1)](../Data/DP1_covid19-peru_x_semanaEpi.csv) de que las primeras dosis recibidas fueron a partir de la SE37 del 2020, posiblemente grupos vulnerables o personas que fueron vacunadas en el extranjero. Debido a que la **vacunación masiva oficialmente comenzó la SE6 del 2021**. De igual manera segundas y terceras dosis en SE9 y SE41 del 2021 respectivamente. *(Hasta aquí todo correcto).*

Sin embargo el problema detectado en [Data Product 6 (DP6)](../Data/DP6_covid19-peru_info-fallecidos.csv) de casos individuales de fallecidos, deja notar que existen registros de personas que recibieron la 1era y 2nda dosis luego de la fecha de fallecimiento. 

**Posibles escenarios:**
- Registros erróneos de fallecimientos *(El más posible)*
- Registros erróneos del id de cada persona
- Registros erróneos en la fecha de vacunación

**Analizando los resultados de DP6:**

```python
import pandas as pd
dp6 = pd.read_csv('../Data/DP6_covid19-peru_info-fallecidos.csv')

# Convertimos en formato fecha las columnas de vacunación
lst_dates = ['fecha_vacuna_1', 'fecha_vacuna_2', 'fecha_vacuna_3']
for date in lst_dates:
    dp6.loc[:,date] = pd.to_datetime(dp6[date], format = '%Y%m%d')
    
dp6 
```

Filtramos todos los casos donde la primera fecha de vacunación sea mayor a la fecha de fallecimiento.

```python
filtered_data = dp6[(dp6['fecha_vacuna_1'] >= dp6['fecha_fallecimiento'])]
filtered_data
```

```python
print(f'Reportes donde la fecha de vacunación es posterior a la fecha de fallecimiento {len(filtered_data)}')
print(f'Representa a {round(len(filtered_data)*100/len(dp6),2)}% del total de fallecidos')
```

- Existen un total de **1646 casos** donde la fecha de vacunación es posterior a la fecha de fallecimiento
- Es decir un **0.81%** del total de fallecidos. 
- **Menos del 1%**

Tomamos 10 casos aleatorios para evaluar

```python
sample = filtered_data.sample(10, random_state = 1).sort_values(by = 'id_persona')
sample = sample[['id_persona', 'fecha_fallecimiento', 'edad', 'dosis_1', 'dosis_2', 'dosis_2',
                 'fecha_vacuna_1', 'fecha_vacuna_2', 'fecha_vacuna_3']]
sample
```

Como se logra apreciar varios de los casos reportados como "fallecidos" tienen hasta 3 dosis recibidas. Por lo que posiblemente haya sido un error registrarlos como fallecimientos.

Sin embargo estos 10 casos aleatorios se los evaluará individualmente para verificar que se traten de la misma persona.

La información personal (especialmente importante el id) se la extrajo del registro de [Fallecidos por COVID-19](https://www.datosabiertos.gob.pe/dataset/fallecidos-por-covid-19-ministerio-de-salud-minsa). Por lo que se procede a comparar con los registros de [Vacunación contra la COVID-19](https://www.datosabiertos.gob.pe/dataset/vacunacion) de acuerdo a su id para verificar que sean la misma persona.

**Se extraen estos 10 id de la muestra aleatoria:**

```python
cases = list(sample.id_persona)
```

**Comparamos el id de cada uno con el id del dataset de vacunación:**

```python
import dask.dataframe as dd
vacunados = dd.read_csv('../RawData/TB_VACUNACION_COVID19.csv', dtype={'edad': 'float64'},
                        usecols = ['id_persona', 'fecha_vacunacion', 'dosis', 'edad'])

vacunados = vacunados[vacunados['id_persona'].isin(cases)].compute().sort_values(by = ['id_persona', 'dosis'])
```
